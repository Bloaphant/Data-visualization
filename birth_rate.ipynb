```python
data = pa.read_csv('worldcountries.csv')
```

```python
loc_countries = [154,54,93,196,68,146,36,149,195,11]
br = data.loc[[154,54,93,196,68,146,36,149,195,11],'Birthrate']
countries = ['Norway', 'Denmark', 'Iceland', 'Switzerland', 'Finland', 'Netherlands', 'Canada', 'New Zealand', 'Sweden', 'Australia']
br_graph = ['11.46','11.13', '13.64', '9.71', '10.45', '10.9', '10.78', '13.76','10.27','12.14']
br.index = countries
rank = list(range(1, 11))
print(br)
```

```python
br = pa.DataFrame(br, dtype='float64')
#dr.astype('float64')
br.plot.bar(figsize=(12,5),width=0.03,facecolor='lightblue')  #figsize to increase the plot scale (in inches), with o mark
plt.xlim([min(arange(10)-0.5),max(arange(10))]) #to give a gap, so it won't touch the x and y axis
plt.ylim(top=20,bottom=0)   #empty spaces on the above
plt.axis(xmax=10)    #empty spaces on the right
rows, cols = br.shape
for col in range(cols):
    for i in arange(10):
        plt.annotate(countries[i],xy=(i-0.3,br.iloc[i,col]+3),fontsize=9)
        plt.annotate(br_graph[i], xy=(i-0.3,br.iloc[i,col]+2), fontsize=9)

plt.xticks(arange(10),rank,rotation=0)
plt.title('Birth rate per 1000 population of top 10 Happiness Rank Countries')
plt.show()
```
