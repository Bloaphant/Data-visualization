```python
data = pa.read_csv('worldcountries.csv')
```

```python
loc_countries = [154,54,93,196,68,146,36,149,195,11]
dr = data.loc[[154,54,93,196,68,146,36,149,195,11],'Deathrate']
countries = ['Norway', 'Denmark', 'Iceland', 'Switzerland', 'Finland', 'Netherlands', 'Canada', 'New Zealand', 'Sweden', 'Australia']
dr_graph = ['9.4','10.36', '6.72', '8.49', '9.86', '8.68', '7.8', '7.53','10.31','7.51']
dr.index = countries
rank = list(range(1, 11))
print(dr)
```

```python
dr = pa.DataFrame(dr, dtype='float64')
#dr.astype('float64')
dr.plot.bar(figsize=(12,5),width=0.03,facecolor='black')  #figsize to increase the plot scale (in inches), with o mark
plt.xlim([min(arange(10)-0.5),max(arange(10))]) #to give a gap, so it won't touch the x and y axis
plt.ylim(top=20,bottom=0)   #empty spaces on the above
plt.axis(xmax=10)    #empty spaces on the right
rows, cols = dr.shape
for col in range(cols):
    for i in arange(10):
        plt.annotate(countries[i],xy=(i-0.3,dr.iloc[i,col]+2),fontsize=9)
        plt.annotate(dr_graph[i], xy=(i-0.3,dr.iloc[i,col]+1), fontsize=9)
plt.xticks(arange(10),rank,rotation=0)
plt.title('Death rate per 1000 population of top 10 Happiness Rank Countries')
plt.show()
```
